```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('batch_size_avg_100.csv')
```

```python
df.plot(figsize=(15,7))
plt.ylabel('Average of Last 100 Episodic Rewards', fontsize=16)
plt.xlabel('Episode', fontsize=16)
plt.title('Average of Last 100 Episodic Rewards For Varying Batch Size', fontsize=16)
plt.xlim(100)
plt.ylim(-150)
plt.savefig('Batch_Size_AVG_100.jpg', bbox_inches='tight')
```

```python
df2 = pd.read_csv('batch_size_epi_avg.csv')
```

```python
df2.plot(figsize=(15,7))
plt.ylabel('Average Episodic Reward over 3 Runs', fontsize=16)
plt.xlabel('Episode', fontsize=16)
plt.title('Episodic Rewards For Varying Batch Size', fontsize=16)
plt.xlim(0)
plt.savefig('Batch_Size_Episodic.jpg', bbox_inches='tight')
```
